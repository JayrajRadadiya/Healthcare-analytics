```python
import io
pcos = pd.read_csv(io.BytesIO(uploaded['Updated PCOS Fertile Data.csv']))
print(pcos)
```

```python
#Drop the unwanted columns
pcos_final =  pcos.drop(['Sl_no','Patient_file_no.'],axis = 1)
```

```python
#replace binary values in column see if there is any effect on the model evaluation
pcos_final["PCOS"] = pcos_final["PCOS"].replace({1:"Yes",0:"No"})
pcos_final["Weight_gain"] = pcos_final["Weight_gain"].replace({1:"Yes",0:"No"})
pcos_final["Hair_growth"] = pcos_final["Hair_growth"].replace({1:"Yes",0:"No"})
pcos_final["Skin_darkening"] = pcos_final["Skin_darkening"].replace({1:"Yes",0:"No"})
pcos_final["Hair_loss"] = pcos_final["Hair_loss"].replace({1:"Yes",0:"No"})
pcos_final["Pimples"] = pcos_final["Pimples"].replace({1:"Yes",0:"No"})
pcos_final["Fast_food"] = pcos_final["Fast_food"].replace({1:"Yes",0:"No"})
pcos_final["Reg_Exe"] = pcos_final["Reg_Exe"].replace({1:"Yes",0:"No"})
pcos_final["Pregnant"] = pcos_final["Pregnant"].replace({1:"Yes",0:"No"})


```

```python
#Creating the BMI category 
def BMI_cat(pcos_final) :
    
    if pcos_final["BMI"] < 19.0 :
        return "Underweight"
    elif (pcos_final["BMI"] > 19.0) & (pcos_final["BMI"] <=25.0):
        return "Normal"
    elif (pcos_final["BMI"]> 25.0) & (pcos_final["BMI"] <= 30.0) :
        return "Overweight"
    elif (pcos_final["BMI"] > 30.0) & (pcos_final["BMI"] < 35.0) :
        return "Obese"
    elif pcos_final["BMI"] > 35.0 :
        return "Extremely Obese"

pcos_final["BMI_cat"] = pcos_final.apply(lambda pcos_final:BMI_cat(pcos_final),
                                      axis = 1)
```

```python
#Making the average column from Number Follicele_no_L & Follicele_no_R 
pcos_final["Avg_Follicele_no"] = ((pcos_final['Follicele_no_L']+pcos_final['Follicele_no_R'])/2)
```

```python
#Similarly creating the Avg length of Follicele columns
pcos_final["Average Follicle Size"] = ((pcos_final['Avg_f_size_L']+pcos_final['Avg_f_size_R'])/2)
```

```python
#Creating an Avg column for Beta Hcg from Beta hcg1 and Beta hcg2
pcos_final["Average_Beta_HCG"] = ((pcos_final['Beta_HCG1']+pcos_final['Beta_HCG2'])/2)
```

```python
# Endometrium value in category
def Endometrium_cat(pcos_final) :
    
    if pcos_final["Endometrium"] < 6.0 :
        return "Under 6mm"
    elif (pcos_final["Endometrium"] >=6.0 ) & (pcos_final["Endometrium"] <= 8.5 ):
        return "6mm - 8.5 mm"
    elif (pcos_final["Endometrium"]> 8.5) & (pcos_final["Endometrium"] <= 11.5) :
        return "9 mm - 11.5 mm"
    elif (pcos_final["Endometrium"] > 11.5) & (pcos_final["Endometrium"] <= 14.0) :
        return "12 mm - 14 mm"
    elif pcos_final["Endometrium"] > 14.0 :
        return "Above 14 mm"

pcos_final["Endometrium_cat"] = pcos_final.apply(lambda pcos_final:Endometrium_cat(pcos_final),
                                      axis = 1)
```

```python
#AMH values in category 
def AMH_cat(pcos_final) :
    
    if pcos_final["AMH"] >= 6.0:
        return "Very High"
    elif (pcos_final["AMH"] >= 3.5 ) & (pcos_final["AMH"] <6.0 ):
        return "Normal"
    elif (pcos_final["AMH"] >= 1.0 ) & (pcos_final["AMH"] <3.5 ):
        return "Optimal"
    elif (pcos_final["AMH"] >= 0.3 ) & (pcos_final["AMH"] <1.0):
        return "Low"
    else:
        return "Very Low"

pcos_final["Amh_cat"] = pcos_final.apply(lambda pcos_final:AMH_cat(pcos_final),
                                      axis = 1)
```
