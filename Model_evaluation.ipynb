**Classification Model-2 Extra Tree Classifier**

```python
#Creating model
classifier_pcos = create_model('et')
```

```python
# Plot of confusion matrix before tuning
plot_model(classifier_pcos, plot = 'confusion_matrix')
```

```python
plot_model(classifier_pcos,plot = 'feature')
```

```python
#Tuning the Extra tree model
tuned_et = tune_model(classifier_pcos)
```

```python
#Evaluating the Extra Tree classifier model (YOu can click on the tiles below to see the model evaluation)
evaluate_model(tuned_et)
```
