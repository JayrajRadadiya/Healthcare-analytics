```python
x=df[[ 'age(yrs)', 'BMI', 'Blood_group', 'Pulse_rate',
       'RR(breaths/min)', 'Hb(g/dl)', 'Cycle_length(days)', 'No_of_abortions', 'Beta_HCG1',
       'FSH/LH','W:H_ratio', 'TSH', 'AMH', 'PRL', 'Vit_D3', 'PRG', 'RBS', 'Weight_gain',
       'Hair_growth', 'Skin_darkening', 'Hair_loss', 'Pimples',
       'BP_sys', 'BP_dias', 'Follicele_no_L', 'Follicele_no_R',
       'Avg_f_size_L', 'Avg_f_size_R', 'Endometrium',"Pregnant"]]
```

### Building a Fully-connected Neural network for predicting PCOS

```python
#Importing supporting elements for a neural network
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers  import Dense,Dropout
from tensorflow.keras.optimizers import RMSprop,SGD
from tensorflow.keras.utils import to_categorical
```

### Scaling Data using Standard Scaler 
#### As we have different of values in dataset we are scaling it to one same format which will me model learn more easily.

```python
#Importing Standard scaler
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
x=ss.fit_transform(x)
x
```

#### As now x is scaled we need to one-hot encode our Target variable as the neural network classifies in that manner

```python
#checking the Shape of our inputs i.e., is x and y
print(x.shape)
print(y.shape)
```

```python
#Spliting out the x into Train and test set
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
y_train=to_categorical(y_train)
```

```python
#Cutting out a little of Training set to Validation set  
x_train,xv_train,y_train,yv_train=train_test_split(x_train,y_train,test_size=0.2,random_state=24)
```

```python
#Building the Model 
#Using Sequential model to build fully-connected Neural Network
#There are total 3 working layer [ Input Layer -- Hidden Layer -- Output Layer]
#Activations used - Relu & softmax
#Number of Neurons - 256,64,2 repectively
model = Sequential([
    Dense(128,activation='sigmoid',input_shape=(30,)),
    Dropout(0.5),
    Dense(64,activation='relu'),
    Dense(2,activation='softmax')
])

model.summary()
```

```python
#Compiling the Model
# Optimizer used : SGD
# Metrics used : Accuracy

op= SGD(0.5)
model.compile(optimizer=op,loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
#Training the model with 100 Epochs
h=model.fit(x_train,y_train,epochs=100,batch_size=50,validation_data=(xv_train,yv_train),verbose=1)
```

```python
print("Minimum Loss on training data : ",round(min(h.history['loss']),3))
print("Maximum Accuracy on training data : ",round(max(h.history['accuracy']),2)*100,"%")
print("Minimum Loss on Validation data : ",round(min(h.history['val_loss']),3))
print("Maximum Accuracy on Validation data : ",round(max(h.history['val_accuracy']),2)*100,"%")
```

```python
import seaborn as sns
```

```python
#Importing confusion matrix from tensorlfow
from tensorflow.math import confusion_matrix 
#Predicting the PCOS for Test Data
y_pred1=model.predict(x_test)
y_pred1=np.argmax(y_pred1,axis=1)
cm=confusion_matrix(y_test,y_pred1)

#import seaborn as sns

ax = sns.heatmap(cm, annot=True, cmap='Blues')

ax.set_title('Confusion Matrix [SGD as Optimizer] \n\n');
ax.set_xlabel('\nPredicted Values')
ax.set_ylabel('Actual Values ');

## Ticket labels - List must be in alphabetical order
ax.xaxis.set_ticklabels(['No PCOS','PCOS'])
ax.yaxis.set_ticklabels(['No PCOS','PCOS'])

## Display the visualization of the Confusion Matrix.
plt.show()
```
